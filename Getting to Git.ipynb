# The Quick Resource to Getting Started with Git

## Starting from Scatch

### These commands will demonstrate how to use Git with Windows (version 2.44.0.windows.1)

#### **Creating a new repository (repo)**
1. Navigate to your git terminal
    - Open command prompt > navigate to your git terminal
    - e.g.) "cd D:\C Junction\GitPortable" > git-cmd
2. check version with: "git --version"
3.  Make an empty directory
    - mkdir "new_directory"  
5. Check the directory to make sure it was created
    - Enter "dir" in the command prompt
6. Navigate to the new directory
    - cd  "new_directory"
7. Initialize the directory
    - git init
8. Place an example file in the directory
    - I've made a html file named "init.html" in Notepad
9. Type "git status" in the terminal to make sure it is present

#### __Git Staging Environment__
1. We now *stage* the file we just created. Enter the following in the cmd
   - git add index.html
   - check that it's been added with "git status"
       - <img src= attachment:e02efd91-31d6-46a9-a451-ba26be9385a3.png style="width:400px"/>
2. Let's test adding multiple files
    - place the updated index.html, bluestyle.css, and README.md into your directory
    - use git add --all
3. Check that all files have been added with "git status"

#### __Git Commit__
- Git considers each commit change point or "save point"

1. git commit -m "your message will go here
2. Using the git log
    -  To view the history of commits for a repository, you can use the log command: git log

#### **Push to git**
1. Make a repo in GitHub
2. Grab the url and enter the command in cmd
    - (git remote add origin url_to_your_project
3. Push all of your changes
    - git push --set-upstream origin master
4. Check GitHub to ensure all changes are present
